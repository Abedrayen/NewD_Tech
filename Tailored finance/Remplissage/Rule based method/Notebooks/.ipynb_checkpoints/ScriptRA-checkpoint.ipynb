```python
from docx import Document

# Charger les données dynamiques
data = {
    "client": {
        "nom": "Dupont",
        "prenom": "Jean",
        "age": 35,
        "residence_fiscale": "France",
        "patrimoine": 150000,
        "sexe": "Homme"
    },
    "investissement": {
        "montant_total": 10000,
        "montant_mensuel": 500
    },
    "objectifs": [
        "Se constituer une épargne",
        "Préparer la retraite",
        "Diversifier le patrimoine"
    ],
    "scpi": {
        "PFO": 4000,
        "PRIMOVIE": 6000
    },
    "esg": {
        "PFO": 30,
        "PRIMOVIE": 20
    }
}

# Fonction pour remplir les paragraphes
def fill_paragraphs(doc, data):
    for paragraph in doc.paragraphs:
        if "Vous souhaitez optimiser vos disponibilités" in paragraph.text:
            paragraph.text = paragraph.text.replace(
                "………………….€",
                f"{data['investissement']['montant_total']} €"
            )
        if "Votre capacité d’épargne est de" in paragraph.text:
            paragraph.text = paragraph.text.replace(
                "……………..€",
                f"{data['investissement']['montant_mensuel']} €"
            )
        if "Monsieur ………………" in paragraph.text:
            paragraph.text = paragraph.text.replace(
                "Monsieur ………………",
                f"Monsieur {data['client']['prenom']} {data['client']['nom']}"
            )
        if "La valeur de votre patrimoine est estimée à environ …………….." in paragraph.text:
            paragraph.text = paragraph.text.replace(
                "……………..€",
                f"{data['client']['patrimoine']} €"
            )

# Fonction pour remplir les tableaux
def fill_tables(doc, data):
    for table in doc.tables:
        # Vérifiez si le tableau a au moins 2 lignes et 2 colonnes
        if len(table.rows) > 1 and len(table.columns) > 1:
            # Exemple de modification pour un tableau spécifique
            if "LE(S) CLIENT(S)" in table.cell(0, 0).text:
                table.cell(1, 0).text = f"Nom : {data['client']['nom']}"
                table.cell(1, 1).text = f"Prénom : {data['client']['prenom']}"
            elif "RAPPEL DE LA DEMANDE" in table.cell(0, 0).text:
                table.cell(1, 0).text = f"Montant total : {data['investissement']['montant_total']} €"
                table.cell(1, 1).text = f"Mensuel : {data['investissement']['montant_mensuel']} €"
        else:
            print("Tableau ignoré (trop petit pour modification).")


# Fonction pour effacer les lignes inutiles
def remove_unnecessary_lines(doc):
    for paragraph in doc.paragraphs:
        if "ATTENTION : Supprimer les lignes non concernées" in paragraph.text:
            p_element = paragraph._element
            p_element.getparent().remove(p_element)

# Charger et remplir le document
def process_document(input_path, output_path, data):
    doc = Document(input_path)
    fill_paragraphs(doc, data)
    fill_tables(doc, data)
    remove_unnecessary_lines(doc)
    doc.save(output_path)
    print(f"Document rempli sauvegardé sous : {output_path}")

# Chemins des fichiers
input_path = "RA SCPI TFI 11-2024.docx"
output_path = "RA_SCPI_TFI_11-2024_Filled.docx"

# Exécuter le traitement
process_document(input_path, output_path, data)
```

```python
from docx import Document

def fill_page_2(doc, data):
    """
    Remplit les informations sur la page 2 : client, consultant, investissement, situation.
    """
    for paragraph in doc.paragraphs:
        if "Nom :" in paragraph.text and "LE(S) CLIENT(S)" in paragraph.text:
            paragraph.text = paragraph.text.replace("Nom :", f"Nom : {data['client']['nom']}")
            paragraph.text = paragraph.text.replace("Prénom :", f"Prénom : {data['client']['prenom']}")
            paragraph.text = paragraph.text.replace("Adresse :", f"Adresse : {data['client']['adresse']}")
            paragraph.text = paragraph.text.replace("E-mail :", f"E-mail : {data['client']['email']}")
        elif "LE CONSEILLER" in paragraph.text:
            paragraph.text = paragraph.text.replace("Nom :", f"Nom : {data['consultant']['nom']}")
            paragraph.text = paragraph.text.replace("Prénom :", f"Prénom : {data['consultant']['prenom']}")
            paragraph.text = paragraph.text.replace("Société :", f"Société : {data['consultant']['societe']}")
        elif "Rappel de la demande" in paragraph.text:
            paragraph.text = paragraph.text.replace("……………", f"{data['investissement']['montant_total']} €")
            paragraph.text = paragraph.text.replace("……………€/mois", f"{data['investissement']['montant_mensuel']} €/mois")
        elif "Situation du client" in paragraph.text:
            paragraph.text = paragraph.text.replace("……………", f"{data['client']['situation']}")

def fill_page_4(doc, data):
    """
    Remplit la page 4 avec l'analyse de la situation : patrimoine et profil investisseur.
    """
    for paragraph in doc.paragraphs:
        if "patrimoine" in paragraph.text.lower():
            paragraph.text = paragraph.text.replace("……………", f"{data['client']['patrimoine']} €")
        if "profil investisseur" in paragraph.text.lower():
            paragraph.text = paragraph.text.replace("1.2.3.4", data['client']['profil_investisseur'])

def fill_page_5(doc, data):
    """
    Remplit la page 5 avec les propositions SCPI.
    """
    scpi_text = "\n".join([f"{scpi} : {montant} €" for scpi, montant in data['scpi'].items()])
    for table in doc.tables:
        try:
            if len(table.rows) > 0 and "Proposition" in table.cell(0, 0).text:
                table.cell(0, 0).text = scpi_text
                total = sum(data["scpi"].values())
                if len(table.rows) == 1:
                    table.add_row()
                table.cell(1, 0).text = f"Montant total de l'investissement : {total} €"
        except IndexError as e:
            print(f"Erreur dans le tableau de proposition : {e}")

def calculate_esg(doc, data, page):
    """
    Calcule et remplit les pourcentages ESG pour chaque SCPI sur la page 17.
    """
    total_investment = sum(data["scpi"].values())
    esg_percentages = {scpi: (amount / total_investment) * 100 for scpi, amount in data['scpi'].items()}

    for table in doc.tables:
        try:
            if len(table.rows) > 0 and "ESG" in table.cell(0, 0).text:
                for scpi, percentage in esg_percentages.items():
                    for row in table.rows:
                        if scpi in row.cells[0].text:
                            if len(row.cells) > 1:
                                row.cells[1].text = f"{percentage:.2f} %"
        except IndexError as e:
            print(f"Erreur dans le calcul ESG : {e}")

def calculate_page_18_19(doc, data, pages):
    """
    Remplit les pages 18 et 19 avec les calculs des pourcentages pour chaque SCPI.
    """
    total_investment = sum(data["scpi"].values())
    for table in doc.tables:
        try:
            if len(table.rows) > 0 and "SCPI" in table.cell(0, 0).text:
                for scpi, montant in data["scpi"].items():
                    for row in table.rows:
                        if scpi in row.cells[0].text:
                            percentage = (montant / total_investment) * 100
                            if len(row.cells) > 1:
                                row.cells[1].text = f"{percentage:.2f} % de {montant} €"
        except IndexError as e:
            print(f"Erreur dans les calculs des pages 18-19 : {e}")

def process_document(input_path, output_path, data):
    """
    Processus principal pour remplir et modifier le document Word.
    """
    doc = Document(input_path)
    fill_page_2(doc, data)
    fill_page_4(doc, data)
    fill_page_5(doc, data)
    calculate_esg(doc, data, 17)
    calculate_page_18_19(doc, data, [18, 19])
    doc.save(output_path)
    print(f"Document rempli sauvegardé sous : {output_path}")

# Exemple de données
data = {
    "client": {
        "nom": "Dupont",
        "prenom": "Jean",
        "adresse": "123 Rue Exemple, Paris",
        "email": "jean.dupont@example.com",
        "patrimoine": 100000,
        "profil_investisseur": "Équilibré (4/7)",
        "situation": "Célibataire"
    },
    "consultant": {
        "nom": "Martin",
        "prenom": "Paul",
        "societe": "Finance Conseil"
    },
    "investissement": {
        "montant_total": 10000,
        "montant_mensuel": 500
    },
    "scpi": {
        "PFO": 5000,
        "PFO2": 3000,
        "Immorente": 2000
    }
}

# Chemins des fichiers
input_path = "RA SCPI TFI 11-2024.docx"
output_path = "RA_SCPI_TFI_11-2024_Filled.docx"

# Exécuter le script
process_document(input_path, output_path, data)
```

```python
from docx import Document

def fill_table(doc, table_index, data):
    """
    Remplit un tableau dans un document Word, même si le tableau est incomplet.
    :param doc: Document Word.
    :param table_index: Index du tableau dans le document.
    :param data: Liste de lignes, chaque ligne étant une liste de colonnes.
    """
    try:
        table = doc.tables[table_index]
        for row_idx, row_data in enumerate(data):
            if row_idx >= len(table.rows):
                print(f"Avertissement : La table {table_index} n'a pas assez de lignes pour l'index {row_idx}.")
                continue

            row = table.rows[row_idx]
            for col_idx, cell_data in enumerate(row_data):
                if col_idx >= len(row.cells):
                    print(f"Avertissement : La ligne {row_idx} de la table {table_index} n'a pas assez de colonnes pour l'index {col_idx}.")
                    continue

                row.cells[col_idx].text = str(cell_data)
    except IndexError:
        print(f"Erreur : Le tableau {table_index} n'existe pas dans le document.")

def fill_document(template_path, output_path, data):
    """
    Remplit un document Word avec des données personnalisées.
    :param template_path: Chemin du modèle de document Word.
    :param output_path: Chemin où enregistrer le document complété.
    :param data: Dictionnaire contenant les données à insérer.
    """
    doc = Document(template_path)

    # Tableau 2: Client et Conseiller
    fill_table(doc, 1, [
        [
            f"Nom : {data['client_name']}",
            f"Nom : {data['advisor_name']}"
        ],
        [
            f"Prénom : {data['client_firstname']}",
            f"Prénom : {data['advisor_firstname']}"
        ],
        [
            f"Adresse : {data['client_address']}",
            f"Société : {data['advisor_company']}"
        ],
        [
            f"E-mail : {data['client_email']}",
            ""
        ]
    ])

    # Tableau 5: Situation du client
    fill_table(doc, 4, [
        [
            data['age'],
            data['profession'],
            data['income'],
            data['children'],
            data['marital_status'],
            data['fiscal_residence'],
            data['risk_profile'],
            data['knowledge_level']
        ]
    ])

    # Tableau 12: Propositions d'investissement
    propositions = [
        [
            "SCPI", "Montant à investir", "Nombre de parts"
        ],
    ] + [
        [prop['name'], f"{prop['amount']} €", prop['parts']] for prop in data['investments']
    ]
    fill_table(doc, 11, propositions)

    # Tableau 22: Calcul ESG
    esg_total = sum(
        prop['amount'] * prop['esg'] / 100 for prop in data['investments']
    )
    total_investment = sum(prop['amount'] for prop in data['investments'])
    esg_percentage = (esg_total / total_investment) * 100 if total_investment > 0 else 0

    fill_table(doc, 21, [
        [
            "TOTAL ALLOCATION", f"{esg_percentage:.2f}%", ""
        ]
    ])

    # Enregistrer le document complété
    doc.save(output_path)

# Exemple de données
example_data = {
    "client_name": "DE LONLAY",
    "client_firstname": "Guillaume",
    "client_address": "",
    "client_email": "",
    "advisor_name": "DE LONLAY",
    "advisor_firstname": "Guillaume",
    "advisor_company": "Tailored Finance",
    "age": 29,
    "profession": "Non Pro",
    "income": "Entre 30 et 70K€",
    "children": "/",
    "marital_status": "Célibataire",
    "fiscal_residence": "France",
    "risk_profile": "Équilibré (3/7)",
    "knowledge_level": "Informé (2/3)",
    "investments": [
        {"name": "PFO", "amount": 15000, "parts": 10, "esg": 30},
        {"name": "PRIMOVIE", "amount": 20000, "parts": 15, "esg": 90},
        {"name": "EURION", "amount": 15000, "parts": 20, "esg": 60}
    ]
}

# Remplir le document
fill_document("RA SCPI TFI 11-2024.docx", "output_filled.docx", example_data)
```

```python
from docx import Document

def fill_table(doc, table_index, data):
    """Remplit un tableau dans un document Word, même s'il est incomplet."""
    try:
        table = doc.tables[table_index]
        for row_idx, row_data in enumerate(data):
            if row_idx >= len(table.rows):
                print(f"Avertissement : La table {table_index} n'a pas assez de lignes pour l'index {row_idx}.")
                continue
            row = table.rows[row_idx]
            for col_idx, cell_data in enumerate(row_data):
                if col_idx >= len(row.cells):
                    print(f"Avertissement : La ligne {row_idx} de la table {table_index} n'a pas assez de colonnes pour l'index {col_idx}.")
                    continue
                row.cells[col_idx].text = str(cell_data)
    except IndexError:
        print(f"Erreur : Le tableau {table_index} n'existe pas dans le document.")

def calculate_esg_percentages(investments):
    """Calcule le pourcentage ESG global basé sur les investissements."""
    total_investment = sum(inv['amount'] for inv in investments)
    esg_total = sum(inv['amount'] * inv['esg'] / 100 for inv in investments)
    esg_percentage = (esg_total / total_investment) * 100 if total_investment > 0 else 0
    return esg_percentage

def fill_document(template_path, output_path, data):
    """Remplit un document Word avec des données personnalisées."""
    doc = Document(template_path)

    # Page 2 : Client, conseiller, et situation
    fill_table(doc, 1, [
        [f"Nom : {data['client_name']}", f"Nom : {data['advisor_name']}"],
        [f"Prénom : {data['client_firstname']}", f"Prénom : {data['advisor_firstname']}"],
        [f"Adresse : {data['client_address']}", f"Société : {data['advisor_company']}"],
        [f"E-mail : {data['client_email']}", ""]
    ])
    fill_table(doc, 4, [[
        data['age'], data['profession'], data['income'], data['children'],
        data['marital_status'], data['fiscal_residence'], data['risk_profile'], data['knowledge_level']
    ]])

    # Page 5 : Propositions SCPI
    propositions = [["SCPI", "Montant à investir (€)", "Nombre de parts"]]
    propositions += [[inv['name'], f"{inv['amount']}", inv['parts']] for inv in data['investments']]
    fill_table(doc, 11, propositions)

    # Page 17 : Calcul ESG
    esg_percentage = calculate_esg_percentages(data['investments'])
    fill_table(doc, 21, [["TOTAL ALLOCATION", f"{esg_percentage:.2f}%", ""]])

    # Pages 18-19 : Calculs spécifiques
    calculations = [["Investissement", "Pourcentage", "Valeur (€)"]]
    for inv in data['investments']:
        calculations.append([inv['name'], "10%", f"{inv['amount'] * 0.1:.2f} €"])
    fill_table(doc, 22, calculations)

    # Enregistrer le document
    doc.save(output_path)
    print(f"Document rempli et enregistré : {output_path}")

# Exemple de données pour remplir le document
example_data = {
    "client_name": "DE LONLAY",
    "client_firstname": "Guillaume",
    "client_address": "123 Rue Exemple",
    "client_email": "guillaume@example.com",
    "advisor_name": "DE LONLAY",
    "advisor_firstname": "Guillaume",
    "advisor_company": "Tailored Finance",
    "age": 29,
    "profession": "Non Pro",
    "income": "Entre 30 et 70K€",
    "children": "/",
    "marital_status": "Célibataire",
    "fiscal_residence": "France",
    "risk_profile": "Équilibré (3/7)",
    "knowledge_level": "Informé (2/3)",
    "investments": [
        {"name": "PFO", "amount": 15000, "parts": 10, "esg": 30},
        {"name": "PRIMOVIE", "amount": 20000, "parts": 15, "esg": 90},
        {"name": "EURION", "amount": 15000, "parts": 20, "esg": 60}
    ]
}

# Remplir le document
fill_document("RA SCPI TFI 11-2024.docx", "RA_SCPI_Filled.docx", example_data)
```

```python
from docx import Document

def flatten_data(data, prefix="", max_enfants=4, max_scpi=5):
    replacements = {}

    if isinstance(data, dict):
        for key, value in data.items():
            new_key = f"{prefix}{key}" if prefix else key
            if isinstance(value, dict):
                replacements.update(flatten_data(value, prefix=new_key + "_"))
            elif isinstance(value, list):
                if "enfants" in new_key:
                    for idx in range(max_enfants):
                        if idx < len(value):
                            replacements.update(flatten_data(value[idx], prefix=f"{new_key}{idx}_"))
                        else:
                            replacements[f"{new_key}{idx}_nom"] = "N/A"
                            replacements[f"{new_key}{idx}_date_naissance"] = "N/A"
                elif "scpi" in new_key:
                    for idx in range(max_scpi):
                        if idx < len(value):
                            replacements.update(flatten_data(value[idx], prefix=f"{new_key}{idx}_"))
                        else:
                            replacements[f"{new_key}{idx}_montant"] = "0 €"
                else:
                    for idx, item in enumerate(value):
                        replacements.update(flatten_data(item, prefix=f"{new_key}{idx}_"))
            else:
                replacements[new_key] = str(value) if value is not None else "N/A"

    return replacements

def calculate_esg_percentage(investments):
    total_investment = sum(item['montant'] for item in investments)
    esg_percentage = sum(item['montant'] for item in investments if item['esg']) / total_investment * 100
    return round(esg_percentage, 2)

def replace_placeholders_in_paragraph(paragraph, replacements):
    full_text = "".join(run.text for run in paragraph.runs)
    for key, value in replacements.items():
        placeholder = f"{{{{{key}}}}}"
        full_text = full_text.replace(placeholder, value)
    if full_text != paragraph.text:
        for run in paragraph.runs:
            run.text = ""
        paragraph.add_run(full_text)

def replace_placeholders_in_table(table, replacements):
    for row in table.rows:
        for cell in row.cells:
            for paragraph in cell.paragraphs:
                replace_placeholders_in_paragraph(paragraph, replacements)

def process_document(template_path, output_path, replacements):
    doc = Document(template_path)
    for paragraph in doc.paragraphs:
        replace_placeholders_in_paragraph(paragraph, replacements)
    for table in doc.tables:
        replace_placeholders_in_table(table, replacements)
    doc.save(output_path)
    print(f"Document saved: {output_path}")

# Example data structure
data = {
    "client_name": "Fabien",
    "client_firstname": "Guillaume",
    "client_age": 29,
    "client_income": "Entre 30 et 70K€",
    "client_status": "Célibataire",
    "client_risk_profile": "Équilibré (3/7)",
    "client_knowledge": "Informé (2/3)",
    "investments": [
        {"nom": "Primovie", "montant": 100000, "esg": True},
        {"nom": "Cœur d’Europe", "montant": 50000, "esg": False}
    ],
    "total_investment": 150000,
    "esg_percentage": 0  # This will be calculated dynamically
}

# Calculate ESG percentage
data["esg_percentage"] = calculate_esg_percentage(data["investments"])

# Paths to the template and output
template_path = "RA SCPI TFI 11-2024.docx"
output_path = "RA_SCPI_Filled.docx"

# Flatten the JSON data
flattened_data = flatten_data(data)

# Fill the document
process_document(template_path, output_path, flattened_data)
```
